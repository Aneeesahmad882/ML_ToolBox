```python
#Answer2
from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Bokeh Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a circle glyph to create a scatter plot
p.circle(x, y, size=10, color="blue", alpha=0.5)

# Add a line glyph to create a line plot
p.line(x, y, line_width=2, line_color="green", legend_label="Line")

# Add bars to create a bar chart
bar_x = [1, 2, 3, 4, 5]
bar_height = [6, 7, 2, 4, 5]
p.vbar(x=bar_x, top=bar_height, width=0.5, color="orange", legend_label="Bars")

# Add patches to create polygons or areas
x_polygon = [1, 2, 3, 4]
y_polygon = [2, 4, 1, 3]
p.patches([x_polygon], [y_polygon], fill_color="purple", fill_alpha=0.6, line_color="black", legend_label="Polygon")

# Show the plot with legends
p.legend.title = 'Glyphs'
show(p)
```

```python
#Answer3

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend, LegendItem

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Bokeh Plot in Google Colab", x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle(x, y, size=10, color="blue", alpha=0.5)

# Customize the title
p.title.text_font_size = "16pt"
p.title.text_font_style = "bold"
p.title.text_color = "green"

# Customize the axes
p.xaxis.axis_label = "Custom X-axis Label"
p.yaxis.axis_label = "Custom Y-axis Label"
p.xaxis.axis_label_text_font_size = "14pt"
p.yaxis.axis_label_text_font_style = "italic"
p.xaxis.axis_label_standoff = 10
p.xaxis.axis_label_text_color = "red"
p.xaxis.axis_line_width = 2
p.yaxis.axis_line_color = "blue"


# Create a legend
legend = Legend(items=[LegendItem(label="Data Points", renderers=[p.circle(x, y, size=10, color="blue", alpha=0.5)])])
p.add_layout(legend, 'above')

# Customize the legend
legend.title = "Legend Title"
legend.label_text_font_size = "12pt"
legend.label_text_font_style = "italic"
legend.label_text_color = "purple"

# Add data to the figure
p.circle(x, y, size=10, color="blue", alpha=0.5, legend_label="Data Points")

# Display the plot in the notebook
show(p)
```

Bokeh is an open-source Python library for creating interactive and visually appealing data visualizations. Bokeh includes a feature called the "Bokeh Server" that allows you to create web applications with real-time interactive plots. With Bokeh Server, you can build data dashboards, live-updating data visualizations, and more.

Here's how you can use Bokeh Server to create interactive plots that can be updated in real time:

Install Bokeh: If you haven't already, you need to install the Bokeh library. You can install it using pip:


#pip install bokeh
Import the necessary modules: In your Python script, import the required Bokeh modules, such as bokeh.plotting, bokeh.models, and bokeh.server.

Create your interactive plot: Define your interactive plot using Bokeh. You can create various types of plots, such as scatter plots, line charts, bar charts, etc.

Define callback functions: To make your plot interactive and update in real time, you'll need to define callback functions. These functions will be triggered when events like button clicks or slider adjustments occur. Callbacks allow you to update the plot's data or properties dynamically.

Create a Bokeh Application: You need to create a Bokeh Application object that encapsulates your interactive plot and the associated callback functions. This is done using bokeh.application.

Run the Bokeh Server: Start the Bokeh Server by running the bokeh serve command in your terminal with the path to your Bokeh application script. For example:


#bokeh serve your_app_script.py
Access the app in a web browser: Once the Bokeh Server is running, you can access your interactive plot by navigating to a URL provided by the Bokeh Server. This URL is usually displayed in the terminal.

Interact with the plot: You can now interact with your plot by using widgets or other interactive elements you've added to your plot, and the plot will update in real time based on the defined callback functions.
